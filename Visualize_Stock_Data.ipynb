```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data  = pd.read_csv("excel_vnm.csv")
data = data[::-1]
price = np.array(data['<CloseFixed>'])
price = price[::-1]
vol   = np.array(data['<Volume>'])
vol   = vol[::-1]
```

```python
fig, ax1 = plt.subplots()

ax1.set_xlabel('TIME')
ax1.set_ylabel('Price', color='tab:red')
ax1.plot(price, color='tab:red')
ax1.tick_params(axis='x')

ax2 = ax1.twinx()
ax2.set_ylabel('Volumn', color='tab:blue')
ax2.plot(vol, color='tab:blue')
ax2.tick_params(axis='y')

fig.tight_layout()
plt.show()
```
